```python
import pandas as pd
```

```python
year = 2021
str = 'https://www.pro-football-reference.com/years/{}/rushing.htm'
```

```python
url2 = str.format(year)
```

```python
df = pd.read_html(url2, header = 1)
```

```python
df
```

```python
rb_df = df[0]
```

```python
df = rb_df.drop(rb_df[rb_df.Player == 'Player'].index) #delete the rows with 'Age' in column 'Age'
rb_df = df[['Player', 'Tm', 'Pos', 'Age', 'Att', 'Yds', 'TD', 'Fmb']]
```

```python
rbdf = rb_df.loc[(rb_df['Pos'] == 'RB')] #locate rows that only contain 
```
