```python
import pandas as pd
```

```python
year = 2021
str = 'https://www.pro-football-reference.com/years/{}/passing.htm'
```

```python
url = str.format(year)
```

```python
df = pd.read_html(url, header = 0)
df
```

```python
# len(df) #this shows how many tables there are, in this case its only 1 
```

```python
df2021 = df[0] #using 0 automatically selects that table 
# df2021[df2021.Age == 'Age'] look for 'Age' in column 'Age'
```

```python
df = df2021.drop(df2021[df2021.Age == 'Age'].index) #delete the rows with 'Age' in column 'Age'
```

```python
qb_df = df[['Player', 'Tm', 'Pos', 'Cmp', 'Att', 'Cmp%', 'Yds', 'TD', 'TD%', 'Int', 'Int%']]
```

```python
qbdf = qb_df.loc[(qb_df['Pos'] == 'qb')] #locate rows that only contain 
qbdf
```
